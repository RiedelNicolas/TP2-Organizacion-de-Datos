```python
import pandas as pd
import numpy as np
from gensim.models import Word2Vec
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv('csv/train.csv')
train.info()
```

```python
tweets = train['text'].str.split()
tweets
```

```python
def cos(x1, x2):
  return np.dot(x1, x2)/(np.linalg.norm(x1)*np.linalg.norm(x2))
```

```python
for i in range(11):
    for j in range(11):
        model = Word2Vec(tweets, min_count=1,size= 100,workers=8, window =3, sg = 1, alpha=0.1*i, min_alpha=0.1*j)
        print("i = " + str(i) + ", j = " + str(j))
        print(cos(model["home"], model["house"]))
        print(cos(model["?"], model["house"]))
        
```

```python
model = Word2Vec(tweets, min_count=1,size= 100,workers=8, window =3, sg = 1, alpha=1, min_alpha=0)
print(cos(model["home"], model["house"]))
model = Word2Vec(tweets, min_count=1,size= 100,workers=8, window =3, sg = 1, alpha=1, min_alpha=0)
print(cos(model["home"], model["house"]))
model = Word2Vec(tweets, min_count=1,size= 100,workers=8, window =3, sg = 1, alpha=1, min_alpha=0)
print(cos(model["home"], model["house"]))
```

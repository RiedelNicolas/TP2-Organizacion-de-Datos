```python
import pandas as pd
import numpy as np
from gensim.models import Word2Vec
from sklearn.manifold import TSNE,SpectralEmbedding,LocallyLinearEmbedding,MDS
from sklearn.decomposition import PCA
import umap
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv('csv/train.csv')
train.info()
```

```python
elmo_train = np.loadtxt("train_embedding_file.txt").reshape(len(train.index),1024)
w2v_train = np.loadtxt("train_embedding_file_w2v(norm).txt").reshape(len(train.index),300)
```

```python
elmo = pd.DataFrame(elmo_train)
elmo.head()
```

# t-SNE

```python
tsne = TSNE(random_state=1342,n_iter=1500,metric='cosine',n_components=2)
tsne.fit(elmo)
```

```python
embd_tr = tsne.fit_transform(elmo)
embd_tr
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(embd_tr[:,0],embd_tr[:,1],hue=train['target'])
```

```python
w2v = pd.DataFrame(w2v_train)
w2v.head()
```

```python
tsne.fit(w2v)
```

```python
embd_tr = tsne.fit_transform(w2v)
embd_tr
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(embd_tr[:,0],embd_tr[:,1],hue=train['target'])
```

```python
todo_junto = pd.concat([w2v,elmo], axis=1)
todo_junto.head()
```

```python
tsne.fit(todo_junto)
```

```python
embd_tr = tsne.fit_transform(todo_junto)
embd_tr
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(embd_tr[:,0],embd_tr[:,1],hue=train['target'])
```

# UMAP

```python
reducer = umap.UMAP()
```

```python
scaled_elmo = StandardScaler().fit_transform(elmo)
```

```python
elmo_umap = reducer.fit_transform(scaled_elmo)
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(elmo_umap[:,0],elmo_umap[:,1],hue=train['target'])
```

```python
scaled_w2v = StandardScaler().fit_transform(w2v)
```

```python
w2v_umap = reducer.fit_transform(scaled_w2v)
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(w2v_umap[:,0],w2v_umap[:,1],hue=train['target'])
```

```python
scaled_together = StandardScaler().fit_transform(todo_junto)
```

```python
ambos = reducer.fit_transform(scaled_together)
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(ambos[:,0],ambos[:,1],hue=train['target'])
```

# PCA

```python
pca = PCA(n_components=2)
```

```python
pca_elmo = pca.fit_transform(scaled_elmo)
plt.figure(figsize=(10,10))
sns.scatterplot(pca_elmo[:,0],pca_elmo[:,1],hue=train['target'])
```

```python
pca_w2v = pca.fit_transform(scaled_w2v)
plt.figure(figsize=(10,10))
sns.scatterplot(pca_w2v[:,0],pca_w2v[:,1],hue=train['target'])
```

```python
pca_ambos = pca.fit_transform(scaled_together)
plt.figure(figsize=(10,10))
sns.scatterplot(pca_ambos[:,0],pca_ambos[:,1],hue=train['target'])
```

# Laplacian Eigenmaps

```python
laplacian = SpectralEmbedding(n_components=2)
```

```python
laplacian_elmo = laplacian.fit_transform(scaled_elmo)
plt.figure(figsize=(10,10))
sns.scatterplot(laplacian_elmo[:,0],laplacian_elmo[:,1],hue=train['target'])
```

```python
laplacian_w2v = laplacian.fit_transform(scaled_w2v)
plt.figure(figsize=(10,10))
sns.scatterplot(laplacian_w2v[:,0],laplacian_w2v[:,1],hue=train['target'])
```

```python
laplacian_ambos = laplacian.fit_transform(scaled_together)
plt.figure(figsize=(10,10))
sns.scatterplot(laplacian_ambos[:,0],laplacian_ambos[:,1],hue=train['target'])
```
